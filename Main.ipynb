```python
def don_limpio(urls):
    resultados = []
    for url in urls:
        response = req.get(url)
        soup = bs(response.text, 'html.parser')
        lista = [i.text for i in soup.find_all("p")]

        indice = None
        for i, char in enumerate(lista):
            if char == "———":
                indice = i

        texto_final = lista[2:indice]
        cuerpo = []
        for str in texto_final:
            cuerpo.append(re.sub(r'[\[\]\{\}]', '', str))

        obra = lista[1]
        resultados.append((cuerpo, obra))

    return resultados
```

### Vamos scrapear Proyect Gutenberg

```python
from tqdm.notebook import tqdm
from collections import Counter
```

```python
def top10(text, n=10, common_words=None):
    if common_words is None:
        common_words = ['the', 'be', 'to', 'of', 'and', 'a', 'in', 'that', 'have', 'I', 'it', 'for', 'not', 'on', 'with', 'he', 'as', 'you', 'do', 'at', 'this', 'but', 'his', 'by', 'from', 'they', 'we', 'say', 'her', 'she', 'or', 'an', 'my', 'would', 'there', 'their', 'what', 'so', 'up', 'out', 'if', 'about', 'who', 'get', 'which', 'go', 'me', 'when', 'make', 'can', 'like', 'no', 'just', 'him', 'know', 'take', 'person', 'into', 'year', 'your', 'good', 'some', 'could', 'them', 'see', 'other', 'than', 'then', 'now', 'look', 'only', 'come', 'its', 'over', 'think', 'also', 'back', 'after', 'use', 'two', 'how', 'our', 'work', 'first', 'well', 'way', 'even', 'new', 'want', 'because', 'any', 'these', 'give', "only","there","them","such", "will","than", "than", "other", "very", "most", "great", "even"]
    w = re.findall('\w+', text.lower())  #la anterior es una lista de palabras demasiado comunes como para aportar información interesante, está incompleta, habría que quitar cuantificadores, por ejemplo.
    word_count = Counter(w)
    for word in common_words:
        del word_count[word]
    return dict(word_count.most_common(n))
```

```python
def extrac_gut(url): #con esta funcion escrapeamos los texto de Proyect Gutenberg, sacando autores, títulos y obras
    res = req.get(url)
    soup = bs(res.text, "html.parser")
    autores = []   
    links_gut = []
    for link in soup.find_all("a", {"class" : "link"}):  
        l = link.get("href")  #en este bucle sacamos los enlaces de cada libro en la página y los guardamos
        if len(l) < 15:
            links_gut.append("https://www.gutenberg.org" + l)

    links_obras = []
    for enlace in links_gut:   
        rec = req.get(enlace)   
        sopa = bs(rec.text, "html.parser") 
        for en in sopa.find_all("a", {"class" : "link"}): 
            l = en.get("href")#entramos a cada enlace de cada libro y buscamos aquellos que estén en formato html5, escrapearemos esos
            if ".html.images" in l:
                try:
                    for name in sopa.find_all("h1"):
                        autores.append(name.text)        
                        links_obras.append("https://www.gutenberg.org" + l)
                except: 
                    pass

    lista_obras = []
    for enla in links_obras:  
        rec = req.get(enla) #conseguimos cada uno de los textos, lo guardamos cada uno en una lista y luego lo juntamos
        sopa = bs(rec.text, "html.parser")
        temp = [i.text for i in sopa.find_all("p")]
        lista_obras.append(temp)

    lis = ["\n", "\r", "\xa0"]             # En el siguiente bucle quitamos caracteres especiales
    pattern = "[" + "".join(lis) + "]"

    for i, lst in enumerate(lista_obras):
        for j, my_str in enumerate(lst):
            cleaned_str = re.sub(pattern, "", my_str) # Quitamos palabras menores de 3 caracteres son ruido a la hora de analizar 
            cleaned_str = re.sub(r'\b\w{1,3}\b', '', cleaned_str)
            lista_obras[i][j] = cleaned_str

    final = {}
    for i, autor in enumerate(autores):
        final[autor] = top10(" ".join(lista_obras[i])) #aplicamos la función para sacar el top 10 palabras más utilizadas

    return final
```

```python
from joblib import Parallel, delayed

BASE_URL = "https://www.gutenberg.org/ebooks/search/?query=philosophy&submit_search=Go%21&start_index={}"

# Crear una lista con los índices de inicio para cada página
start_indices = [i for i in range(26, 526, 25)]

# Crear un conjunto de procesos en paralelo para extraer datos de todas las páginas
with tqdm(total=len(start_indices)) as pbar:
    results2 = Parallel(n_jobs=6)(delayed(extrac_gut)(BASE_URL.format(start_index)) for start_index in start_indices)
    pbar.update()
```
