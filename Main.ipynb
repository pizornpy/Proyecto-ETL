```python
def don_limpio(url):    
    response = req.get(url)
    soup = bs(response.text, 'html.parser')
    lista = [i.text for i in soup.find_all("p")]
    
    
    indice = None
    for i, char in enumerate(lista):
        if char == "———":
            indice = i

    texto_final = lista[2:indice]
    cuerpo = []
    for str in texto_final:
        cuerpo.append(re.sub(r'[\[\]\{\}]', '', str))
    
    cuerpo
    obra = lista[1]
    return cuerpo, obra 
```

```python
def don_limpio2(urls):
    resultados = []
    for url in urls:
        response = req.get(url)
        soup = bs(response.text, 'html.parser')
        lista = [i.text for i in soup.find_all("p")]

        indice = None
        for i, char in enumerate(lista):
            if char == "———":
                indice = i

        texto_final = lista[2:indice]
        cuerpo = []
        for str in texto_final:
            cuerpo.append(re.sub(r'[\[\]\{\}]', '', str))

        obra = lista[1]
        resultados.append((cuerpo, obra))

    return resultados
```
